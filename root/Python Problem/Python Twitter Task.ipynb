### Importing All Necessarry Python Libraries

```python
import tweepy
import jsonlines
import pandas as pd
```

### Add the Consumer Keys and Access Tokens

```python
consumer_key = "XXXXXXXXXXXXXXXXX"
consumer_secret = "XXXXXXXXXXXXXXXXXXXXXX"
access_token = "XXXXXXXXXXXXXXXXXXXXXXXX"
access_token_secret = "XXXXXXXXXXXXXXXXXXXXXXXXX"
```

```python
# Creating the tweepy authentication object.
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)

# Setting your access token and access secret.
auth.set_access_token(access_token, access_token_secret)

# Creating the API object while passing in auth information.
api = tweepy.API(auth) 
```

```python
#UserName of Twitter Account
name = "midasIIITD"

# Number of tweets to retrieve
tweetCount = 75

# Calling the user_timeline function with our parameters
results = api.user_timeline(id=name, count=tweetCount)
```

```python
#Using JSONLines in Write Mode to dump Response in jsonl file.
with jsonlines.open("results.jsonl",mode='w') as writer:
    for tweet in results:
        writer.write(tweet._json)
```

```python
#Using JSONLines to parse the results file.
answer=[]
with jsonlines.open("results.jsonl",mode='r') as reader:
    for item in reader.iter():
        temp=[]
        temp.append(item['text'])
        temp.append(item['created_at'][:16])
        temp.append(item['favorite_count'])
        temp.append(item['retweet_count'])
        temp.append(item['entities'])
        answer.append(temp)

#Creating a DataFrame using Pandas
df=pd.DataFrame(answer,columns=['text','data','likes','retweets','entities'])
```
